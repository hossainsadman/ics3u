# Task

Your task is to create a weighted average and grade point average calculator for students.

Get a series of course marks out of `100`. In addition, get the
course weight, either `1.0` or `0.5`. Exit when `-1` is entered or the user has entered `5.0` or more courses.

Calculate the *weighted average* course mark rounded to 1 decimal place.

Calculate the *grade point average* according to this scheme:

> If the weighted average is: \
> 90 to 100: `4.0` \
> 80 to  less than 90: `3.8` \
> 70 to less than 80: `3.4` \
> 60 to less than 70: `2.8` \
> 50 to less than 60: `2.0` \
> less than 50: `1.0` \


Note: The weighted average of numbers is the sum of the products of the numbers and their weights, divided by the total weight, eg. if the numbers were `80`, `77`, and `92`, with weights `1.0`, `0.5`, and `0.5`, respectively, then the weighted average would be:

`(80*1.0 + 77*0.5 + 92*0.5) / (1 + 0.5 + 0.5)
= 82.25`

Create a simple, friendly, and professional user interface. Display the results (total course weight, weighted average, grade point average) to the user. No need to validate inputs. Minimal comments required. Use concepts learned in class.

Write the code below.

```
EXIT_MARK = -1
EXIT_WEIGHT = 5.0
# constant FW represents # of decimal places to show for final output
FW = 1

course_num = 1
sum_courses = 0
total_weight = 0
weighted_avg = 0
gpa = 1.0

while True:
    
    # does not print a new line if mark & weight for first 
    # course is being inputted
    if course_num != 1:
        print()

    course_mark = int(input("Course {} Mark: ".format(course_num)))
    if course_mark == EXIT_MARK :
        break
    
    course_weight = float(input("Course {} Weight: ".format(course_num)))
    if course_weight >= EXIT_WEIGHT:
        break
    
    # calculations below only take place if inputs for course mark & weight
    # above are valid
    total_weight += course_weight
    sum_courses += course_mark * course_weight
    course_num += 1

# calculate weighted average 
weighted_avg = sum_courses / total_weight

# calculate grade point average if it is not less than 50
if 50 <= weighted_avg < 60:
    gpa = 2.0
elif 60 <= weighted_avg < 70:
    gpa = 2.8
elif 70 <= weighted_avg < 80:
    gpa = 3.2
elif 80 <= weighted_avg < 90:
    gpa = 3.8
elif 90 <= weighted_avg <= 100:
    gpa = 4.0

# display final totals
print()
print("Total Course Weight: {}".format(total_weight))
print("Weighted Average: {:.{width}f}".format(weighted_avg, width = FW))
print("Grade Point Average: {}".format(gpa))
```
