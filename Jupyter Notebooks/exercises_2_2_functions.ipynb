1. Return the speed of an object given the distance travelled (`m`) and the time (`s`), e.g. `12` metres in `2.5` seconds returns `4.8`.

````
def speed(m, s):
    """Returns speed of an object given distance travelled (m) and the time (s)"""
    return m / s

print(speed(12, 2.5))
````

2. Return the number of digits present in a string given as parameter, e.g `'high5'` returns `1`



````
def num_of_digits(string):
    """Returns the number of digits present in a string (parameter)"""
    counter = 0
    for character in string:
        if character in "0123456789":
            counter += 1
    return counter

print(num_of_digits("high560"))
````

3. Return the sum of the digits of a string grouped as 2-digit numbers. If there is an odd number of digits, the last digit will be a single, eg. `'12952'` returns `109` (since `12 + 95 + 2 = 109`)

````
def sum_of_digits(string):
    """Returns sum of digits of a string that is grouped into 2 digit numbers"""
    sum = 0
    for i in range(0, len(string), 2):
        sum += int(string[i:i + 2])
    return sum

print(sum_of_digits("12952"))
````

4. Return the sum of the digits of a string grouped as `n`-digit numbers, where `n` is a positive integer, eg. `'129527'` in `3`s returns `656` (since `129 + 257 = 656`), and in `4`s returns `1322` (since `1295 + 27 = 1322`).

````
def sum_of_digits(string, n):
    """Returns sum of digits of string (string) grouped as (n)-digit numbers"""
    sum = 0
    for i in range(0, len(string), n):
        sum += int(string[i:i + n])
    return sum

print(sum_of_digits("129527", 4))
````

````
def string_separator(string):
    """Prints a string (string) so each character is on its own line preceded by its index"""
    for i in range(0, len(string)):
        print("index {}: \'{}\'\n".format(i, string[i]))

string_separator("spam")
````

6. Write a function `to_upper()` which converts all lowercase letters to uppercase and returns the new string, and a function `to_lower()` which does the opposite.

````
def to_upper(string):
    """Returns input string with all lowercase letters converted to uppercase"""
    upper_string = ""
    for letter in string:
        if 97 <= ord(letter) <= 122:
            upper_string += chr(ord(letter) - 32)
        else:
            upper_string += letter
    return upper_string

def to_lower(string):
    """Returns input string with all uppercase letters converted to lowercase"""
    lower_string = ""
    for letter in string:
        if 65 <= ord(letter) <= 87:
            lower_string += chr(ord(letter) + 32)
        else:
            lower_string += letter
    return lower_string

print(to_upper("aBc 12"))
print(to_lower("aBc12"))
````

7. Print the first `n` Fibonacci numbers nicely with commas, eg. 
```
The first 6 Fibonacci numbers are:
1, 1, 2, 3, 5, 8, ...
```

````
def fibonacci(n):
    """Prints the first (n) fibonacci numbers"""
    temp_1 = 1
    temp_2 = 1
    for i in range(n):
        if i % 2 == 0:
            print(temp_1, end = ", ")
            temp_1 += temp_2
        elif i % 2 != 0:
            print(temp_2, end = ", ")
            temp_2 += temp_1
        
fibonacci(6)
````

8. Write a function `max_(a, b)` which returns the larger of the numbers `a` and `b`, or `a` if they are equal. Also write a similar function `min_(a, b)`. These are basically copies of the built-ins `max()` and `min()`.

````
def max_(a, b):
    """Returns the larger of the numbers (a) and (b)"""
    if b > a:
        return b
    else:
        return a

def min_(a, b):
    """Returns the smaller of the numbers (a) and (b)"""
    if a > b:
        return b
    else:
        return a

print(max_(1, 2))
print(min_(1, 2))
````

9. Three positive integers $a$, $b$, and $c$ with $a < b < c$ form a *Pythagorean Triple* if $a^2 + b^2 = c^2$.  For example `(3, 4, 5)` is a Pythagorean Triple since $3^2 + 4^2 = 5^2$. Write a function that takes a positive integer parameter and then finds and prints all Pythagorean Triples whose largest member is less than or equal to that integer. [Challenge: do not print Pythagorean Triples that are multiples of other triples, e.g. `(6, 8, 10)` is simply `2 * (3, 4, 5)`.]
Eg.
```
Enter a positive integer: 14
Pythagorean Triples:
(3, 4, 5)
(6, 8, 10)
(5, 12, 13)
```

````
from math import sqrt

def pytha_triple(c):
    """Prints all pythagorean triples whose largest member is less than or equal
       to that integer"""

    for hyp in range(c + 1):
        for b in range(1, hyp + 1):
            a = (sqrt(hyp**2 - b**2))
            if a % 1 == 0 and a != 0:
                print("({}, {}, {})".format(hyp, b, a))
        
pytha_triple(14)
````

10. The *Caesar Cipher* is a simple encryption technique where each letter of the alphabet is shifted left or right a certain amount, eg.
```
plaintext character:  ABCDEFGHIJKLMNOPQRSTUVWXYZ
ciphertext character: FGHIJKLMNOPQRSTUVWXYZABCDE
```
In the above example, `A` was shifted `5` places so that it would be encrypted as `F`. A sample message `HELLO` would then be encrypted as `MJQQT`. Write a function that takes a positive integer key (between `1` and `25`) and a string, and encrypts it using the Caesar Cipher. (Traditionally, all messages are converted to uppercase to ease in encryption/decryption.) [Challenge: change the ciphertext to always be output in blocks of 5 uppercase letters, such as `JFCFE XREUK YREBJ WFIRC CKYVW ZJY` (Can you decrypt this?)]

````
def caesar_encrypt(string, n):
    encrypted = ""
    string = to_upper(string)
    for letter in string:
        if letter in " ":
            encrypted += " "
        else:
            if ord(letter) + n > 90:
                ascii = ord(letter) - 25 - 1 + n
            else: 
                ascii = ord(letter) + n
            encrypted += chr(ascii)
    return encrypted

def caesar_decrypt(string, n):
    decrypted = ""
    string = to_upper(string)
    for letter in string:
        if letter in " ":
            decrypted += " "
        else:
            if ord(letter) - n < 65:
                ascii = ord(letter) + 25 + 1 - n
            else: 
                ascii = ord(letter) - n
            decrypted += chr(ascii)
    return decrypted

print(caesar_encrypt("ABCDEFGHIJKLMNOPQRSTUVWXYZ", 5))
print(caesar_decrypt("FGHIJKLMNOPQRSTUVWXYZABCDE", 5))
print(caesar_decrypt("JFCFE XREUK YREBJ WFIRC CKYVW ZJY", 17))
print()
for i in range(25 + 1):
    print(i, end = ": ")
    print(caesar_decrypt("JFCFE XREUK YREBJ WFIRC CKYVW ZJY", i))
````
