```
# 1. Write a function my_min(my_list: list) which finds and returns the smallest 
# value of a list.
def my_min(my_list: list):
  smallest = my_list[0]
  for item in my_list:
    if item < smallest:
      smallest = item
  return smallest

# 2. Write a function that uses linear search to search a list for a value given 
# as parameter, e.g. search(my_list: list, searched_value) -> bool.
def search(my_list: list, searched_value) -> bool:
  for item in my_list:
    if item == searched_value:
      return True
  return False

# 3. a) Write a function that finds the largest value of a list and prints the 
# value and its index. Parameter: a list.
def maximum(my_list: list):
  largest_index = 0
  largest = my_list[0]
  for i in range(1, my_list):
    if my_list[i] > largest:
      largest_index = i
      largest = my_list[i]
  print("Largest value:", largest)
  print("At index:", largest_index)

# 3. b) Write a function as in (a) but for minimum.
def minimum(my_list: list):
  smallest_index = 0
  smallest = my_list[0]
  for i in range(1, my_list):
    if my_list[i] < smallest:
      smallest_index = i
      smallest = my_list[i]
  print("Smallest value:", smallest)
  print("At index:", smallest_index)

# 4. Write a function that takes two parameters: a new value, and a sorted list. 
# The function should insert the new value into the list at the appropriate location 
# so that the list remains sorted. For testing, only insert numbers into numeric 
# lists, and strings into string lists. (No append and sort!)
def insert(sorted_list: list, value) -> list:
  index = len(sorted_list) + 1
  for i in range(len(sorted_list)): 
    if sorted_list[i] > value:
      index = i 
      break
  new_list = sorted_list[:index] + [value] + sorted_list[index:]
  return new_list

# 5. a) Write a function that uses linear search to search a list for a value and 
# counts and returns the number of occurrences of that value. Parameters: the 
# searched value, the list. For testing, obviously, create a list that has 
# repeated values.
def linear_search(my_list: list, value) -> list:
  counter = 0
  for item in my_list:
    if item == value:
      counter += 1
  return counter

# 5. b) Write a function similar to part (a), but instead returns the index of the 
# first occurrence only of the searched value. If the value is not in the list, 
# return -1. Therefore, your function should stop searching once the value has 
# been found!
def first_index_of(my_list: list, value) -> list:
  for i in range(len(my_list)):
    if my_list[i] == value:
      return i
  return -1

def main():
  print(insert([1,2,3,4], 2.5))
  print(linear_search([1,2,2,3,3,3],2))
  print(first_index_of([1,2,3,2],4))
  
if __name__ == '__main__':
  main()
```
