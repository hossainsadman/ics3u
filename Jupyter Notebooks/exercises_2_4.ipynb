```
def to_celsius(fahr_temp: float) -> float:
    """Return the cel_temp in fahrenheit after converting from celsius

    >>>cel_temp(32)
    0
    >>>cel_temp(14)
    -10
    """
    
    fahr_temp = (cel_temp - 32) * 5 / 9
    return fahr_temp
```

```
def get_hypotenuse_len(side1: float, side2: float) -> float:
    """Return length of hypotenuse given legs side1 and side2 of a triangle

    >>>get_hypotenuse_len(3,4)
    5
    >>>get_hypotenuse_len(15,20)
    25
    """
    
    from math import sqrt
    return sqrt(side1**2 + side2**2)
```
