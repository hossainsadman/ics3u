1. Create a dictionary of at least 5 key-value pairs. Use for-loops to:
* print all the keys.
* print all the values.
* print all the key-value pairs on their own line nicely like this: `key : value`





```
a = {'hello': 5, 'world': 5, 'this': 4, 'is': 2, 'the': 3, 'end': 3}
print(a)

# print keys
print('\nkeys:')
for k in a.keys():
    print(k)

# print values
print('\nvalues:')
for v in a.values():
    print(v)

# print key-value pairs
print('\nkey-value pairs')
for k, v in a.items():
    print('{} : {}'.format(k, v))

```

2. a) Create an empty dictionary. Ask the user to enter a key followed by a value. Add these to the dictionary. If the user enters an empty line (e.g. an empty string), stop and then print the dictionary. (This will result in a dictionary of `str: str`.)

b) Convert the code to a function called get_dict() which returns the dictionary.


```
def get_dict() -> dict:
    """Return a dictionary of string and string pairs gotten from the user."""

    a = {}
    while True:
        key = input('Enter a key: ')
        if key != '':
            value = input('Enter a value for key "{}": '.format(key))
            a[key] = value
        else:
            break
    return a


a = get_dict()
print(a)
```

3. Create a simple translation dictionary that has a word in English as key and it's translation in another language as value. Start with at least three words. Make a user interface which asks the user for  a word to translate. If it's in the dictionary, print the translation; if it doesn't exist, prompt the user to add the word and its translation to the dictionary. E.g. `en_fr = {'cat': 'chat', 'cheese': 'fromage', 'parrot': 'perroquet'}`

```
en_fr = {'cat': 'chat', 'cheese': 'fromage', 'parrot': 'perroquet'}

print('English-French Translation Dictionary')

while True:
    en = input('What word in English do you want to translate: ')
    if en in en_fr:
        print('English: {}\n French: {}'.format(en, en_fr[en]))
    else:
        print('Not in the dictionary.')
        fr = input('Enter the French translation for {}: '.format(en))
        en_fr[en] = fr
        print('Success!')
        print('English: {}\n French: {}'.format(en, en_fr[en]))
    print()

```

4. Write a program to count the occurrences of each unique character in a string using a dictionary named `count`. You should loop through the string and add characters as keys. The values should be the number of times the character occured. At the end, you should print the dictionary. Hint: Use the `setdefault()` method to ensure that the key is in the count dictionary (with a default value of `0`) so the program doesnâ€™t throw a `KeyError` error when you first encounter the character.
Eg. The string `'massasauga rattlesnake'` should produce the dictionary
 `{'m': 1, 'a': 6, 's': 4, 'u': 1, 'g': 1, ' ': 1, 'r': 1, 't': 2, 'l': 1, 'e': 2, 'n': 1, 'k': 1}`


```
def count_char_freq(text: str) -> dict:
    """Return a dictionary of the frequency count of unique characters in the
    string text.

    >>> count_char_freq('aba')
    {'a': 2, 'b': 1}
    >>> count_char_freq('massasauga rattlesnake')
    {'m': 1, 'a': 6, 's': 4, 'u': 1, 'g': 1, ' ': 1, 'r': 1, 't': 2, 'l': 1, 'e': 2, 'n': 1, 'k': 1}
    """

    count = {}
    for char in text:
        # Add the char to the dictionary, with freq 0; ignored if key exists (seen).
        count.setdefault(char, 0) 
        # Then increment by 1.
        count[char] += 1
    return count

a = count_char_freq('aba')
b = count_char_freq('massasauga rattlesnake')

print(a)
print(b)
```
