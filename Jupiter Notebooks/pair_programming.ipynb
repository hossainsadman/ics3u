```
#  EXERCISE LEAP YEAR

#  this is imported to use the time.sleep() function, which will be explained later.
import time
#  helps ensure the program runs over again, in a loop.  
while True:

  #  get year input from user
  year = int(input("\nEnter a year: "))

  #  if year is divisible by 4 but not 100 OR year is divisible by 400,
  #  then the inputted year is a leap year
  if year % 4 == 0 and year % 100 != 0 or year % 400 == 0:
    leap_or_not = "a"
  #  if year is not divisible by 4 or not divisible by 400,
  #  inputted year is not a leap year
  else:
    leap_or_not = "not a"

  #  display whether inputted year is a leap year
  print("The year {} is {} leap year.".format(year, leap_or_not))
  #  runs a delay between the start and end times pf the program
  time.sleep(1.2)
```

```
#  EXERCISE Line Segement Facts
#  import square root function
from math import sqrt

#  request user for x & y-values of 2 points
print("Enter the coordinates of two points.")

#  get x & y-values of 2 points from user
x1 = float(input("Point 1 x-value: "))
y1 = float(input("Point 1 y-value: "))
x2 = float(input("Point 2 x-value: "))
y2 = float(input("Point 2 y-value: "))

#  calculate length of line formed between inputted points
line_length = sqrt((x2 - x1)**2 + (y2 - y1)**2)

#  display inputted points to user rounded to whole numbers
print("\nYour points: ({:.0f}, {:.0f}) and ({:.0f}, {:.0f})".format(x1, y1, x2, y2))

text = "of the line segment"

#  if x-values of inputted points are not equal, slope is not undefined
if x1 != x2:
  slope = (y2 - y1) / (x2 - x1)
  print("\nThe slope {} is {:.2f}.".format(text, slope))
#  if x-values of inputted points are equal, slope is undefined
else:
  print("\nThe slope {} is undefined.".format(text))

#  display length of line segment formed between input 
print("The length {} {:.2f}.".format(text, line_length))
```
