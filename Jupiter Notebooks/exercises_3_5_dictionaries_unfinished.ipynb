```
dict = {'1':'a','2':'b','3':'c','4':'d','5':'e'}

print('Keys:')
for key in dict.keys():
    print(key)

print('\nValues:')
for value in dict.values():
    print(value)

print('(\'Key\', \'Value\')')
for tuple in dict.items():
    print(tuple)

print('\nKey : Value')
for key in dict.keys():
    print(key, ':', dict[key])

print('\nKey : Value')
for key, value in dict.items():
    print('Key: {}, Value: {}'.format(key, value))
```

```
dict = {}

while True:
    key = input("Enter a key: ")
    if key == '' or key.isspace():
        break
    value = input('Enter a value for key: ')
    dict.update({key: value})

print('\nUser Dictionary:')
print('-'*len('User Dictionary'))
for k,v in dict.items():
    print('Key: {}, Value: {}'.format(k,v))
```

```
def get_dict():
    dict = {}

    while True:
        key = input("Enter a key: ")
        if key == '' or key.isspace():
            break
        value = input('Enter a value for key: ')
        dict.update({key: value})
    
    return dict

def print_dict(msg: str, my_dict: dict):
    print('{}:'.format(msg))
    print('-'*len(msg))
    for k,v in my_dict.items():
        print('Key: {}, Value: {}'.format(k,v))

dicti = get_dict()
print_dict('\nMy Dictionary', dicti)
```
