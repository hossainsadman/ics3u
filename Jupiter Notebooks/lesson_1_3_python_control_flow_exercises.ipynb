```
#  Exercise A

#  get 2 integer inputs from users
num1 = int(input("Number 1: "))
num2 = int(input("Number 2: "))

#  check if inputted integers are equal
equality = ""
if num1 == num2:
  equality = "equal"
else:
  equality = "not equal"
#  display whether numbers are equal
print("The numbers are {}.".format(equality))
```

```
#  Exercise B

#  get integer input from user
num = int(input("Enter a number: "))

#  check which intervals the inputted integer is within
text = ""
if 0 < num < 100:
  text = "high"
elif -100 < num < 0:
  text = "low"
elif num == 0:
  text == "zero"
else:
  text == "outside"
#  display which interval the inputted integer is within
print(text)
```

```
#  Exercise C

#  get integer input from user
num = int(input("Enter a number: "))

#  check if inputted integer is odd or even
text = ""
if num % 2 == 0 and not num == 0:
  text = "even"
elif not num % 2 == 0:
  text = "odd" 
print(num, "is an", text, "number.")
```

```
#  Exercise E

num1 = int(input("Number 1: "))
num2 = int(input("Number 2: "))

larger_num = 0
num_order = ""
if num1 > num2:
  larger_num = num1
  num_order = "first"
else:
  larger_num = num2
  num_order = "second"
print("The larger number is the {} number inputted: {}".format(num_order, larger_num))
```

```
#  Exercise F

a1 = float(input("Angle 1: "))
a2 = float(input("Angle 2: "))
a3 = float(input("Angle 3: "))

t_type = ""
if a1 + a2 + a3 != 180:
   t_type = "not possible"
else:
  if a1 == a2 and a1 == a3:
    t_type = "equilateral"
  elif (a1 == a2 != a3) or (a2 == a3 != a1) or (a1 == a3 != a2):
    t_type = "isosceles"
  else:
    t_type = "scalene"
print("Triangle is {}.".format(t_type))
```
