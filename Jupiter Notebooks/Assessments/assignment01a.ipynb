```
# Question 2: Sphere Facts

# import math module containing sqrt function and pi constant
from math import sqrt, pi

# print program title
print("Sphere Facts")
# get surface area of sphere from user
surface_area = float(input("Enter the surface area: "))

radius = sqrt((surface_area/(4*pi)))
volume = (4/3)*pi*(radius**3)

# print radius and volume of sphere
print("\nThe radius of the sphere is {}, and the volume is {}.".format(radius, volume))
```

## Question 3: Pythagoras

Write a program that asks the user for the lengths of one leg of a right triangle and the hypotenuse (`float`). Calculate the length of the other leg and also the area. Print the statement using `.format()`. (Useful formulae: *a*Â² + *b*Â² = *c*Â², *A* = 1/2 *bh*).


The program output should look *exactly* like this:

\

```
# Question 3: Pythagoras

# import math module containing square root function
from math import sqrt

# print program title
print("Pythagoras")
# get length of first leg from user
leg_1 = float(input("Enter the first length: "))
# get length of hypotenuse from user
hypotenuse = float(input("Enter the hypotenuse: "))

leg_2 = sqrt(hypotenuse**2 - leg_1**2)
area = 0.5*leg_1*leg_2

# print length of second leg of right triangle
print("\nThe other leg is {}.".format(int(leg_2)))
# print area of right triangle
print("The area is {}.".format(int(area)))
```
