1. Get a string from the user; print the string with the first and last 
characters removed, eg. input `'Hello!'`, output `'ello'`. Use `for` loop and slice.

```
new_text = ""
new_text2 = ""

text = input("Enter a string: ")

# use slice to obtain substring of input inclusively 
# extending from second char (index 1) to second 
# last char (index of length of input - 1)
new_text = text[1:-1]

# for loop iterates from second char (index 1) to 
# second last char (index of length of input - 1)
for i in range(1, len(text) - 1):
    # in each iteration, the char is concatenated 
    # to the new string
    new_text2 += text[i]
    
print(new_text)
print(new_text2)
```

2. Get a string from the user; print the string with the first and last character swapped, eg. input `'Hello!'`, output `'!elloH'`. Use `for` loop and slice.



```
new_text = ""
new_text2 = ""

text = input("Enter a string: ")

# concatenate substrings from input of second last char 
# (index of string length - 1), of second char to second 
# last char (index 1 to index of string length - 1), and 
# of first char (index 0)
new_text = text[-1] + text[1:-1] + text[0]
new_text3 = text[::-1]

# concatenate last char of string to become index 0 of
# empty string
new_text2 = text[-1]
# for loop iterates from second char (index 1) to 
# second last char (index of length of input - 1)
for i in range(1, len(text) - 1):
    # in each iteration, the char is concatenated 
    # to the new string
    new_text2 += text[i]
# concatenate first char of string to become last index 
# of new substring
new_text2 += text[0]

print(new_text)
print(new_text2)
print(new_text3)
```

3. Print all of the ASCII characters from `!` to `~`. Even better, print them nicely formatted in a table with their decimal value beside them, similar to the one shown in the lesson.

```
# constants declare start & end values for ASCII table, and field width for table
START = 33
END = 126
WIDTH = 5

print("Deci. | Char.")
print("-" * (WIDTH * 2 + 3))
for i in range(START, END + 1):
    print("{:{width}} | {}".format(i, chr(i), width = 5))
```

4. Get a string from the user; make and print a new string which has every third letter of the input string, starting from the second, e.g. input `'Hello World!'`, output `'eood'`. Use `for` loop and slice.

```
new_text = ""
new_text2 = ""

text = input("Enter a string: ")

# for loop iterates every 3rd character from index 1 (second character) to last 
# index of string 
for i in range(1, len(text), 3):
    # character at iterated indices are concatenated to a new string
    new_text += text[i]

new_text2 = text[1::3]

print(new_text)
print(new_text2)
```

5. Get a string from the user; make and print a new string which is the input string reversed, eg. input `'cheese'`, output `'eseehc'`. Use `for` loop (hint: create an empty string and add each character to it in the order desired [two ways to do this]) and slice. 

```
new_text = ""
new_text2 = ""
new_text3 = ""

text = input("Enter a string: ")

# for loop iterates each character starting from last index (same value as length
# of string) to first index (0 inclusive is same as -1 exclusive)
for i in range(len(text) - 1, -1, -1):
    # concatenate each character in string to new string
    new_text += text[i]

# in addition of strings, order of operations matters (e.g. a + b != b + a)
for i in range(len(text)):
    new_text3 = text[i] + new_text3

# slice reverses string
new_text2 = text[::-1]

print(new_text)
print(new_text2)
```

6. Get input from the user. Make a new string which has changed all lowercase letters into uppercase letters from the input string. **Do not change any other characters.** Use the ASCII table and `ord()` and `chr()` only. Print the new string, eg. input `'Hello there!'`, output `'HELLO THERE!'`. (`.upper()` is not allowed.)

```
new_text = ""

text = input("Enter a string: ")

# for loop iterates through each character in string
for char in text:
    # if decimal value of character corresponds to a value of a lowercase letter,
    # the character corresponding to the value of that uppercase letter (is 32 
    # units less than lowercase letter) is concatenated to new string
    if 97 <= ord(char) <= 122:
        new_text += chr(ord(char) - 32)
    # all non-lowercase letters are concatenated to new string
    else:
        new_text += char
        
print(new_text)
```

7. Do the opposite of #6: uppercase to lowercase. (Of course, `.lower()` is not allowed.)

```
new_text = ""

text = input("Enter a string: ")

# for loop iterates through each character in string
for char in text:
    # if decimal value of character corresponds to a value of a uppercase letter,
    # the character corresponding to the value of that uppercase letter (is 32 
    # units higher than lowercase letter) is concatenated to new string
    if 65 <= ord(char) <= 90:
        new_text += chr(ord(char) + 32)
    else:
        new_text += char

print(new_text)
```

8. Get the sum of the digits of an integer string given by the user (i.e. a *string* that contains only characters `0123456789`) by converting each digit to `int` and summing them, e.g. input `'154698'`, output `33`.

```
number = 0

text = input("Enter an integer: ")

# for loop iterates through each character in string
for digit in text:
    # each character in string is converted to integer value & concatenated to 
    # new string
    number += int(digit)

print(number)
```

9. Do the same as #8, except using two-digit numbers, e.g input `'321476'`, output `122` (since `32 + 14 + 76 = 122`). If there is an odd number of digits, the final digit is taken alone.

```
number = 0

text = input("Enter an integer: ")

# for loop iterates from first index (0) to second last index (if string lenth
# is odd) or last index (if string length is even)
for i in range(0, len(text), 2):
    # concatenate every pair of integers to string;  with odd numbers, slice does
    # not give error but rather goes to charcter index
    number += int(text[i:i + 2])
print(number)
```
